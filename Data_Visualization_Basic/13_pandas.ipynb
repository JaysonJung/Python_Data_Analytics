```python
# 필요 라이브러리 임포트
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
import matplotlib as mpl
import seaborn as sns
# Jupyter Notebook에서 Matplotlib 출력을 위한 설정
%matplotlib inline
# 한글 사용을 위한 설정
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
# Matplotlib 출력 크기 설정
plt.rcParams["figure.figsize"] = (14,4)
plt.rcParams['lines.linewidth'] = 2
plt.rcParams['lines.color'] = 'r'
plt.rcParams['axes.grid'] = True 
```

```python
np.random.seed(0)
df_data = pd.DataFrame(np.random.randn(100, 3), index=pd.date_range('1/1/2018', periods=100), columns=['A', 'B', 'C']).cumsum()
df_data.head()
```

```python
df_data.plot()
plt.title("판다스")
plt.xlabel("시간")
plt.ylabel("데이터")
plt.show()
```

`plot` 메서드의 `kind` 인수

* `bar`
* `pie`
* `hist`
* `kde`
* `box`
* `scatter`
* `area`

```python
iris = sns.load_dataset("iris")    # 붓꽃 데이터
titanic = sns.load_dataset("titanic")    # 타이타닉호 데이터
```

```python
iris.sepal_length[:20].plot(kind='bar', rot=0)
plt.title("막대 그래프")
plt.xlabel("데이터")
plt.ylabel("꽃받침 길이")
plt.show()
```

`kind` 인수에 문자열을 쓰는 대신 `plot.bar`처럼 직접 메서드로 사용할 수도 있다.

```python
iris[:5].plot.bar(rot=0)
plt.title("Iris 데이터의 Bar Plot")
plt.xlabel("Data")
plt.ylabel("각 Feature의 값")
plt.ylim(0, 7)
plt.show()
```

```python
df2 = iris.groupby(iris.species).mean()
df2.columns.name = "feature"
df2
```

```python
df2.plot.bar(rot=0)
plt.title("각 종의 Feature별 평균")
plt.xlabel("평균")
plt.ylabel("종")
plt.ylim(0, 8)
plt.show()
```

```python
df2.T.plot.bar(rot=0)
plt.title("각 Feature의 종별 평균")
plt.xlabel("Feature")
plt.ylabel("평균")
plt.show()
```

다음은 ``pie``,  ``hist``, ``kde``, ``box``, ``scatter`` 플롯 예제이다.

```python
df3 = titanic.pclass.value_counts()
df3.plot.pie(autopct='%.2f%%')
plt.title("선실별 승객 수 비율")
plt.axis('equal')
plt.show()
```

```python
iris.plot.hist()
plt.title("각 Feature 값들의 빈도수 히스토그램(Histogram)")
plt.xlabel("데이터 값")
plt.show()
```

```python
iris.plot.kde()
plt.title("각 Feature 값들의 빈도수에 대한 커널밀도그림(Kernel Density Plot)")
plt.xlabel("데이터 값")
plt.show()
```

```python
iris.plot.box()
plt.title("각 Feature 값들의 빈도수에 대한 상자그림(Box Plot)")
plt.xlabel("Feature")
plt.ylabel("데이터 값")
plt.show()
```

박스플롯에 대해서는 추가적인 기능을 가진 `boxplot` 명령이 별도로 있다.

* `boxplot`: https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.boxplot.html

```python
iris.boxplot(by='species')
plt.title("각 Feature의 종별 데이터에 대한 상자그림(Box Plot)")
plt.show()
```

```python
iris.plot.scatter(x='sepal_length', y='sepal_width')
plt.title("꽃받침의 길이와 너비에 대한 산점도(Scatter Plot)")
plt.show()
```
